# Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
data['Rating'].dtype
```

```python
data['Rating'].sort_values().unique()
```

```python
def rating_classification(Rating):
    if Rating >=1 and Rating<3:
        return 'Average'
    if Rating >=3 and Rating<6:
        return 'Good'
    else:
        return 'Excellent'
```

```python
data['Rating Classification'] = data['Rating'].apply(rating_classification)
```

# Find Unique Values From Genre 

```python
data['Genre'].unique()
```

```python
len(data['Genre'].unique())
```

```python
data["Genre"].value_counts()
```

```python
list_1 = [] # to check unique values in genre first we have to split the values 
for values in data['Genre']:
    list_1.append(values.split(','))
```

```python
list_1
```

```python
one_d= [] # then we will assign the splitted values in another empty list

for items in list_1:
    for item in items:
        one_d.append(item)
```

```python
one_d
```

```python
uni_list = []
for items  in one_d:
    if items not in uni_list:
        uni_list.append(items)
```

```python
uni_list
```
